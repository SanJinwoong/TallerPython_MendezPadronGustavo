```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('datos.xlsx')
df.head()
```

```python
df.describe()
```

```python
print(df['Parcial 1'].mean())
print(df['Parcial 1'].min())
print(df['Parcial 1'].max())
```

```python
df['promedio'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)
df['estatus'] = np.where(df['promedio'] >= 7.5, 'Aprobado', 'Reprobado')
df
```

```python
alumno = 5
notas=df[df['ID']==alumno][['Parcial 1', 'Parcial 2', 'Parcial 3']].iloc[0]

plt.plot(['Parcial 1', 'Parcial 2', 'Parcial 3'], notas)
plt.ylim(0, 10)
plt.ylabel(f'Notas del alumno {alumno}')
plt.title('Evaluaciones de alumnos')
plt.show()
```
