## Sección 1: Visualización de Datos con Matplotlib

**Tema:** Gráficos de Líneas - Relación entre Variables

Creación de gráficos para visualizar la relación entre dos variables continuas.

```python
# Importación de la librería Matplotlib para crear visualizaciones
import matplotlib.pyplot as plt

# Definición de datos: mediciones de temperatura a lo largo del tiempo
tiempo = [1, 4, 6, 8, 10, 12]  # Tiempo en horas
temperatura = [20, 23, 27, 25, 23, 19]  # Temperatura en grados Celsius

# Creación del gráfico de líneas
plt.plot(tiempo, temperatura)
plt.title("Temperatura vs Tiempo")
plt.xlabel("Tiempo (horas)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)  # Activar cuadrícula para mejor lectura
plt.show()
```

### Gráficos de Barras

**Objetivo:** Visualizar datos categóricos - Comparación de promedios por materia.

```python
# Gráfico de barras para comparar promedios de diferentes materias
materias = ['Programación', 'Física', 'Geometría', 'Cálculo']
promedio = [66, 55, 78, 92]

# Creación del gráfico de barras
plt.bar(materias, promedio)
plt.title("Promedio por Materia")
plt.xlabel("Materias")
plt.ylabel("Promedio")
plt.grid(True, axis='y')  # Cuadrícula solo en eje Y para mejor lectura
plt.show()
```

## Sección 2: Análisis Numérico con NumPy

**Tema:** Operaciones con Arrays y Estadística Descriptiva

NumPy es la librería fundamental para cálculo numérico en Python.

```python
# Importación de NumPy para operaciones numéricas eficientes
import numpy as np

# Creación de un array con mediciones de temperatura
temps = np.array([20, 23, 27, 25, 23, 19])

# Cálculo de estadísticas descriptivas básicas
print('Temperaturas registradas:', temps)
print('Promedio:', temps.mean())
print('Temperatura mínima:', temps.min())
print('Temperatura máxima:', temps.max())
```

### Operaciones Vectorizadas

**Concepto:** NumPy permite realizar operaciones elemento a elemento de forma eficiente.

```python
# Cálculo de potencia eléctrica usando la Ley de Ohm (P = V * I)
# Arrays con valores constantes de voltaje y variaciones de corriente
voltajes = np.array([5, 5, 5, 5])
corrientes = np.array([1, 2, 3, 4])

# Multiplicación elemento a elemento (operación vectorizada)
potencias = voltajes * corrientes
print('Potencias calculadas (Watts):', potencias)
```

### Histogramas de Distribución

**Objetivo:** Visualizar la distribución de frecuencias de un conjunto de datos.

```python
# Creación de histograma para visualizar la distribución de calificaciones
# Dataset con calificaciones de estudiantes
calificaciones = np.array([60, 70, 45, 67, 89, 34, 23, 56, 65, 98])

# Generación del histograma con 5 bins (intervalos)
plt.hist(calificaciones, bins=5, edgecolor='black')
plt.title("Distribución de Calificaciones")
plt.xlabel("Rangos de Calificación")
plt.ylabel("Frecuencia")
plt.show()
```
